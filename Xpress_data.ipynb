## s+r demand rate.txt

```python
demand_s = pd.read_csv(PATH1+'DEMAND_RATE_PREDICTION_TABLE_suppliers.csv', header=0, engine='python', encoding='utf-8-sig', index_col=0)
demand_s
```

```python
demand_s = demand_s.sort_index()
demand_s = demand_s[:15]
demand_s
```

```python
demand_r = pd.read_csv(PATH1+'DEMAND_RATE_PREDICTION_TABLE_receivers.csv', header=0, engine='python', encoding='utf-8-sig', index_col=0)
demand_r
```

```python
demand_r.sort_index(inplace=True)
demand_r
```

```python
demand = demand_s.append(demand_r)
demand
```

```python
demand.to_csv(path+'demand.txt', index=False, header=None, sep="\t")
demand.to_csv(path+'demand.csv', index=True)
```

```python
merge = pd.merge(state, demand, left_index=True, right_index=True)
merge
```

```python
merge['dif'] = merge['current_inventory']-merge['rate_pred']*4
merge
```

```python
positive = merge['dif'][merge['dif']>=0]
positive.sum()
```

```python
negative = merge['dif'][merge['dif']<=0]
negative.sum()
```
